```python
import os
from os import listdir
from os.path import isfile, join
mypath = './dataset'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
X = {}
Y = []
LBS = 12
for file in onlyfiles:
    f = open('./dataset/' + file)

    x_ = []
    for l in f:
        x = np.array([item.rstrip() for item in l.split()])
        x.astype(np.float)
        x_.append(x)
    X.update({file: np.array(x_).reshape([-1, 64, 80]).astype(np.float32)})
    f.close()
```

```python
plt.imshow(X['Rec--000013_AmpAll.dat'][13])
plt.show()
plt.imshow(X['Rec--000075_AmpAll.dat'][3])
plt.show()
plt.imshow(X['Rec--000098_AmpAll.dat'][3])
plt.show()
```

```python
# ok
from sklearn.cluster import KMeans
ims = [X['Rec--000013_AmpAll.dat'][13], X['Rec--000075_AmpAll.dat'][3], X['Rec--000098_AmpAll.dat'][3]]
clusters_data = {}
i = 0
for item in ims:
    vals = item.reshape([-1,1])
    model = KMeans(n_clusters=3)
    # fit the model
    model.fit(vals)
    # assign a cluster to each example
    yhat = model.predict(vals)
    # retrieve unique clusters
    clusters = np.unique(yhat)
    clusters_data.update({'cluster_' + str(i): yhat, 'cluster_centers_' + str(i): model.cluster_centers_})
    # create scatter plot for samples from each cluster
    plt.imshow(yhat.reshape([64, 80]))
    plt.show()
    i += 1
```
